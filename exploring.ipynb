```python
col_names =  list(df.columns) # sets list of column names
ordered_col_names = [] # empty list
i = 0 
lowest = col_names[0] # initalize lowest value, random  does not matter if actually lowest

while len(col_names) > 0:
    # compares each column name with lowest
    # if less than lowest sets as new lowest value
    # uses string.lower() method for consiststency
    if col_names[i].lower() < lowest.lower():
        lowest = col_names[i]
    i += 1
    # restarts loop after appending/removing 1st lowest value
    if i == len(col_names):
        ordered_col_names.append(lowest)
        col_names.remove(lowest)
        if col_names:
            lowest = col_names[0]
        i = 0

        
new_df = pd.DataFrame() # intializes new empty dataframe

# creates new dataframe column and sets its values 
for colname in ordered_col_names:
    new_df[colname] = df[colname]
    
new_df.head()
```
